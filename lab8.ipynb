## 4. –ò–º–ø–ª–µ–º–µ–Ω—Ç–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import ParseError

BCCD_DIR = "BCCD"
IMAGE_SIZE = (416, 416)
GRID_SIZE = 13
NUM_BOXES = 1
NUM_CLASSES = 3

class SimpleDetector(nn.Module):
    def __init__(self):
        super(SimpleDetector, self).__init__()

        self.backbone = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),

            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.1),
            nn.MaxPool2d(2, 2),
        )

        self.detection_head = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.1),
            nn.Conv2d(512, (5 + NUM_CLASSES) * NUM_BOXES, kernel_size=1, stride=1, padding=0),
        )

    def forward(self, x):
        x = self.backbone(x)
        x = self.detection_head(x)
        x = x.permute(0, 2, 3, 1)
        return x

class BCCDDataset(Dataset):
    def __init__(self, image_dir, annotation_dir, transform=None):
        self.transform = transform
        self.image_dir = image_dir
        self.annotation_dir = annotation_dir
        self.image_files = [f for f in os.listdir(self.image_dir) if f.endswith('.jpg')]

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_name = self.image_files[idx]
        img_path = os.path.join(self.image_dir, img_name)
        annotation_path = os.path.join(self.annotation_dir, img_name.replace('.jpg', '.txt'))

        image = Image.open(img_path).convert('RGB')
        width, height = image.size

        target = torch.zeros((GRID_SIZE, GRID_SIZE, 5 + NUM_CLASSES))

        f = open(annotation_path, 'r')
        for line in f:
            class_id, x_center, y_center, box_width, box_height  = map(float, line.strip().split())
            class_id = int(class_id)

            grid_x = int(x_center * GRID_SIZE)
            grid_y = int(y_center * GRID_SIZE)

            target[grid_y, grid_x, 0] = 1
            target[grid_y, grid_x, 1:5] = torch.tensor([x_center, y_center, box_width, box_height])
            target[grid_y, grid_x, 5 + class_id] = 1

        if self.transform:
            image = self.transform(image)

        return image, target


def detection_loss(pred, target):
    pred_boxes = pred[..., :5]
    pred_classes = pred[..., 5:]

    target_boxes = target[..., :5]
    target_classes = target[..., 5:]

    obj_mask = target_boxes[..., 0] == 1

    box_loss = nn.MSELoss()(pred_boxes[obj_mask], target_boxes[obj_mask])

    class_loss = nn.BCEWithLogitsLoss()(pred_classes[obj_mask], target_classes[obj_mask])

    total_loss = box_loss + class_loss
    return total_loss
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
transform = transforms.Compose([
    transforms.Resize(IMAGE_SIZE),
    transforms.ToTensor(),
])

dataset = BCCDDataset(image_dir='/content/drive/MyDrive/Colab Notebooks/multimedia/datasets/BCCD/images/train',
                      annotation_dir='/content/drive/MyDrive/Colab Notebooks/multimedia/datasets/BCCD/labels/train',
                      transform=transform)
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)

model_cust = SimpleDetector()


optimizer = optim.Adam(model_cust.parameters(), lr=0.001)
```

–û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

```python
from tqdm import tqdm

num_epochs = 3
for epoch in range(num_epochs):
    model_cust.train()
    running_loss = 0.0

    for images, targets in tqdm(dataloader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]'):
        optimizer.zero_grad()

        outputs = model_cust(images)
        loss = detection_loss(outputs, targets)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {running_loss / len(dataloader)}")

torch.save(model_cust.state_dict(), "simple_detector.pth")
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏

```python
import numpy as np
from collections import defaultdict

def calculate_iou(box1, box2):
    """
    –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç IoU (Intersection over Union) –º–µ–∂–¥—É –¥–≤—É–º—è bounding box'–∞–º–∏
    box: [x_center, y_center, width, height] (–Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã)
    """
    box1 = [
        box1[0] - box1[2]/2,  # x1
        box1[1] - box1[3]/2,  # y1
        box1[0] + box1[2]/2,  # x2
        box1[1] + box1[3]/2   # y2
    ]
    box2 = [
        box2[0] - box2[2]/2,
        box2[1] - box2[3]/2,
        box2[0] + box2[2]/2,
        box2[1] + box2[3]/2
    ]

    x_left = max(box1[0], box2[0])
    y_top = max(box1[1], box2[1])
    x_right = min(box1[2], box2[2])
    y_bottom = min(box1[3], box2[3])

    if x_right < x_left or y_bottom < y_top:
        return 0.0

    intersection_area = (x_right - x_left) * (y_bottom - y_top)

    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1])
    union_area = box1_area + box2_area - intersection_area

    return intersection_area / union_area

def evaluate_model(model, dataloader, device, iou_threshold=0.5):
    model.eval()

    metrics = defaultdict(list)
    class_names = ['WBC', 'RBC', 'Platelets']

    with torch.no_grad():
        for images, targets in dataloader:
            images = images.to(device)
            targets = targets.to(device)

            outputs = model(images)

            for i in range(images.shape[0]):
                pred_boxes, pred_classes = process_output(outputs[i], conf_threshold=0.5)
                true_boxes, true_classes = process_target(targets[i])

                matched = set()

                for pred_idx, (pred_box, pred_class) in enumerate(zip(pred_boxes, pred_classes)):
                    best_iou = 0
                    best_true_idx = -1

                    for true_idx, (true_box, true_class) in enumerate(zip(true_boxes, true_classes)):
                        if true_idx in matched:
                            continue

                        iou = calculate_iou(pred_box, true_box)
                        if iou > best_iou and iou >= iou_threshold and pred_class == true_class:
                            best_iou = iou
                            best_true_idx = true_idx

                    if best_true_idx != -1:
                        matched.add(best_true_idx)
                        metrics[class_names[pred_class]].append(1)
                    else:
                        metrics[class_names[pred_class]].append(0)

                for true_idx in range(len(true_boxes)):
                    if true_idx not in matched:
                        metrics[class_names[true_classes[true_idx]]].append(-1)

    results = {}
    for class_name in class_names:
        if class_name not in metrics:
            continue

        tp = metrics[class_name].count(1)
        fp = metrics[class_name].count(0)
        fn = metrics[class_name].count(-1)

        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        results[class_name] = {
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'tp': tp,
            'fp': fp,
            'fn': fn
        }

    aps = []
    for class_name in class_names:
        if class_name in results:
            aps.append(results[class_name]['precision'])
    map_score = np.mean(aps) if aps else 0

    results['mAP'] = map_score

    return results

def process_output(output, conf_threshold=0.5):
    """
    –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –≤—ã—Ö–æ–¥ –º–æ–¥–µ–ª–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç bounding boxes –∏ –∫–ª–∞—Å—Å—ã
    """
    boxes = []
    classes = []

    for grid_y in range(GRID_SIZE):
        for grid_x in range(GRID_SIZE):
            cell_data = output[grid_y, grid_x]

            if cell_data[0] >= conf_threshold:
                x_center = cell_data[1].item()
                y_center = cell_data[2].item()
                width = cell_data[3].item()
                height = cell_data[4].item()

                class_probs = cell_data[5:].softmax(dim=0)
                class_id = torch.argmax(class_probs).item()

                boxes.append([x_center, y_center, width, height])
                classes.append(class_id)

    return boxes, classes

def process_target(target):
    """
    –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç —Ü–µ–ª–µ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏—Å—Ç–∏–Ω–Ω—ã–µ bounding boxes –∏ –∫–ª–∞—Å—Å—ã
    """
    boxes = []
    classes = []

    for grid_y in range(GRID_SIZE):
        for grid_x in range(GRID_SIZE):
            cell_data = target[grid_y, grid_x]

            if cell_data[0] == 1:
                x_center = cell_data[1].item()
                y_center = cell_data[2].item()
                width = cell_data[3].item()
                height = cell_data[4].item()

                class_id = torch.argmax(cell_data[5:]).item()

                boxes.append([x_center, y_center, width, height])
                classes.append(class_id)

    return boxes, classes
```
